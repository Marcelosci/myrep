####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Supondo que 'sinasc' já esteja carregado da Tarefa 1
# sinasc = pd.read_csv('sinasc_RO_2019.csv')

# 1) Boxplot da idade da mãe
plt.figure(figsize=(6, 4))
sns.boxplot(y=sinasc['IDADEMAE'])
plt.title('Boxplot da Idade da Mãe')
plt.ylabel('Idade da Mãe')
plt.show()
```

####  2) Faça um boxplot do peso do bebê

```python
# 2) Boxplot do peso do bebê
plt.figure(figsize=(6, 4))
sns.boxplot(y=sinasc['PESO'])
plt.title('Boxplot do Peso do Bebê')
plt.ylabel('Peso do Bebê (gramas)')
plt.show()
```

####  3) Faça um histograma da idade da mãe

```python
# 3) Histograma da idade da mãe
plt.figure(figsize=(8, 5))
sns.histplot(sinasc['IDADEMAE'], kde=True, bins=20) # bins=20 é um bom ponto de partida, pode ajustar
plt.title('Histograma da Idade da Mãe')
plt.xlabel('Idade da Mãe')
plt.ylabel('Frequência')
plt.show()
```

#### 4) Faça um histograma do peso do bebê

```python
# 4) Histograma do peso do bebê
plt.figure(figsize=(8, 5))
sns.histplot(sinasc['PESO'], kde=True, bins=30) # Ajuste bins conforme a distribuição
plt.title('Histograma do Peso do Bebê')
plt.xlabel('Peso do Bebê (gramas)')
plt.ylabel('Frequência')
plt.show()
```
