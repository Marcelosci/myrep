#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Carregar a base
sinasc = pd.read_csv('sinasc_RO_2019.csv')

# Verificar as primeiras linhas
sinasc.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
plt.figure(figsize=(15, 10))

# APGAR1
plt.subplot(2, 3, 1)
sinasc['APGAR1'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição de APGAR1')
plt.xlabel('APGAR1')
plt.ylabel('Frequência')

# APGAR5
plt.subplot(2, 3, 2)
sinasc['APGAR5'].value_counts().sort_index().plot(kind='bar')
plt.title('Distribuição de APGAR5')
plt.xlabel('APGAR5')
plt.ylabel('Frequência')

# SEXO
plt.subplot(2, 3, 3)
sinasc['SEXO'].value_counts().plot(kind='bar')
plt.title('Distribuição por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Frequência')

# GESTACAO
plt.subplot(2, 3, 4)
sinasc['GESTACAO'].value_counts().plot(kind='bar')
plt.title('Distribuição de Gestação')
plt.xlabel('Semanas de Gestação')
plt.ylabel('Frequência')

# ESTCIVMAE
plt.subplot(2, 3, 5)
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar')
plt.title
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.figure(figsize=(8,6))
sns.barplot(
    x=sinasc['CONSULTAS'].astype(str), 
    y=sinasc['APGAR5'], 
    estimator='mean', 
    ci=None
)
plt.title('APGAR5 médio por Número de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('APGAR5 Médio')
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.figure(figsize=(14, 10))

# SEXO
plt.subplot(2, 2, 1)
sinasc['SEXO'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribuição por Sexo')
plt.xlabel('Sexo')
plt.ylabel('Frequência')

# GESTACAO
plt.subplot(2, 2, 2)
sinasc['GESTACAO'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Distribuição de Gestação')
plt.xlabel('Semanas de Gestação')
plt.ylabel('Frequência')

# ESTCIVMAE
plt.subplot(2, 2, 3)
sinasc['ESTCIVMAE'].value_counts().plot(kind='bar', color='salmon')
plt.title('Estado Civil da Mãe')
plt.xlabel('Estado Civil')
plt.ylabel('Frequência')

# CONSULTAS
plt.subplot(2, 2, 4)
sinasc['CONSULTAS'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Distribuição de Consultas')
plt.xlabel('Número de Consultas')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()
```
